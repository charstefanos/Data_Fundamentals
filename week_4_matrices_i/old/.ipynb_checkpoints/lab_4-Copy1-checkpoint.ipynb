### C. 

Complete the function below, that given a reference distillery name and a norm, will compute:
* the distance to every distillery to the reference distillery, as a vector
* the *angle* between each distillery vector and the reference distillery, using the inner product and L2 norm.

```python
def whisky_map(distillery, norm):
    """
    Parameters:
        distillery: name of the distillery to compare to
        norm: p-norm to use for distances
    Returns:
        distances: a vector of distances using the given p-norm
        angles: angles between each vector and the given distillery
    """
    
    ### BEGIN SOLUTION
    distance = np.linalg.norm(whisky-distillery, norm, axis=0)
    angles = []
    for w in whisky:
        # cos theta = a.b / |a||b|
        cos_theta = np.dot(w, distillery) / (np.linalg.norm(distillery)*np.linalg.norm(w))
        # now we just invert the cosine
        angle = np.arccos(cos_theta)
        angles.append(angle)
            
    ### END SOLUTION
    return distances, angles
    
```
