## Vector quantisation 

Vector quantisation is a widely used approach in image and video compression. It represents sub-patches of images (or videos, or sounds, or whatever) as vectors, then finds a small set of representative vectors (**cluster centres**) that represent the image reasonably well. The image can then be reconstructed from just those few vectors, rather than transmitting or storing all of the image data.

#### B. 
The code below will take a matrix of vectors (one vector per row) and return a matrix of **cluster centres** of those vectors. There is a 2D example shown below to show how to use the clustering function. It works just the same for any number of dimensions.

Use this code to find 64 cluster centres for `img`, and generate a new image where each 16x16 patch of the original image has been replaced with its nearest cluster centre. The image should be called `reconstructed_img` and should be viewable using `ia.show_image()`. It should look like the original image, but with lower quality. You will need to use `img_tensor` for this calculation.

Note: you will need to **think** about how to do this. 

```python
from sklearn.cluster import KMeans

def cluster(x, clusters):
    # return the cluster centers for an NxD matrix of D-dimensional vectors
    # return will be clusters x D dimensional matrix    
    means = KMeans(clusters)
    means.fit(x)
    return means.cluster_centers_
```

```python
# Clustering demo in 2D

# create some 2D data in three distinct clusters, 150 points total
x = np.random.normal(0,1, (50,2))
x = np.concatenate([x, np.random.normal(0,1, (50,2)) + [3, 5]])
x = np.concatenate([x, np.random.normal(0,1, (50,2)) + [-6, 2]])

# find the cluster centres
centers  = cluster(x, 3)

# plot the data and the centers
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x[:,0], x[:,1], s=20, label="Data")
ax.scatter(centers[:,0], centers[:,1], s=50, c='r', label="Cluster centers")
ax.legend()
```

```python
# reshape to 1024x256 array (each row is one patch)
img_array = np.einsum('ijk -> kij', img_tensor).reshape(256, 1024).T
# cluster to 64 clusters
clusters = cluster(img_array, 64)

# reconstruct by finding the **nearest** patch
# in the cluster centre list
reconstructed = []

# there is a vectorised way to do this
# but it isn't very memory efficient
for row in img_array:
    reconstructed.append(clusters[np.argmin(np.linalg.norm(row-clusters, axis=1))])

# stack it back
# the reshaping here is a bit tricky, but it is really just straightforward
# einsumming
reconstructed_array = np.stack(reconstructed)    
re = reconstructed_array.reshape(32,32,16,16)
reconstructed_image = np.einsum('ijkl -> iljk', re).reshape(512,512).T
```

#### C.
Complete the function below, to perform all of these steps on any image and return a new image with the cluster replacement applied. This should be a small number of lines of simple code [the reference solution is 12 lines, excluding comments]. If you have written a massive program, you are doing it wrong.

Your function only needs to work with values of `size` that are exact multiples of the image dimensions. You don't have to solve the case when it doesn't divide evenly.

```python
def vq(img, size, clusters):
    """
    Parameters:
        img: 2D image
        size: size of the patches to split the image into
        clusters: number of clusters to find
        
    Returns:
        The image reconstructed from the clustered vectors found.
    """
    # this is just  copy/pasting the solution above
    # and then making sure the sizes are set from the parameters/image size
    n_row = img.shape[0] // size
    n_col = img.shape[1] // size
    img_tensor = np.einsum('ijkl->kijl',
                           img.reshape(n_row, size, n_col, size)).reshape(
                               n_row, n_col, size * size)
    img_array = np.einsum('ijk -> kij', img_tensor).reshape(
        size * size, n_row * n_col).T
    clusters = cluster(img_array, clusters)

    reconstructed = []
    for row in img_array:
        reconstructed.append(clusters[np.argmin(np.linalg.norm(row - clusters, axis=1))])

    reconstructed_array = np.stack(reconstructed)
    re = reconstructed_array.reshape(n_row, n_col, size, size)
    reconstructed_image = np.einsum('ijkl -> iljk', re).reshape(img.shape[0], img.shape[1]).T
    return reconstructed_image
```
